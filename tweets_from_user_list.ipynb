```python
import requests
import json
import pandas as pd


def get_user_tweets(username):
    
    url = f"https://syndication.twitter.com/srv/timeline-profile/screen-name/{username}"

    r = requests.get(url)

    html = r.text

    start_str = '<script id="__NEXT_DATA__" type="application/json">'
    end_str = '</script></body></html>'

    start_index = html.index(start_str) + len(start_str)
    end_index = html.index(end_str, start_index)

    json_str = html[start_index: end_index]
    data = json.loads(json_str)

    l = []
    for i in range(len(data["props"]["pageProps"]["timeline"]["entries"])):
        l.append({
            'author':username,
            'date':data["props"]["pageProps"]["timeline"]["entries"][i]["content"]["tweet"]["created_at"],
            'text':data["props"]["pageProps"]["timeline"]["entries"][i]["content"]["tweet"]["full_text"],
            'retweet_count': data["props"]["pageProps"]["timeline"]["entries"][i]["content"]["tweet"]["retweet_count"]
            }
        )
    df = pd.DataFrame(l)
    df['date_clean']= pd.to_datetime(df.date)
    df.sort_values('date_clean')

    return df

def get_all_tweets():
    df_list =[]
    users_list = ['GoogleAI', 'HeyAbhishekk','geoffreyhinton', 'ylecun','AIatMeta']
    for user in users_list:
        df = get_user_tweets(user)
        df_list.append(df)
    df = pd.concat(df_list)
    df = df[['retweet_count','author','text','date_clean']]
    df['retweet_count']=df['retweet_count'].astype(int)
    df = df.sort_values('retweet_count', ascending=False)
    return df
```

```python
df = get_all_tweets()
```

```python
df.drop_duplicates()[:20].to_csv('data/50_tweets.csv', index=False)
```
